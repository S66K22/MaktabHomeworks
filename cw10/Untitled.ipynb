```python
import numpy as np
import matplotlib.pyplot as plt

#  
np.random.seed(42)

#  
centers = [
    [0, 0],
    [3, 3],
    [0, 4]
]

#        
cov = np.array([
    [2.0, 1.2],
    [1.2, 2.0]
])

n_samples_per_cluster = 300

#  
X = []
y = []

for i, center in enumerate(centers):
    samples = np.random.multivariate_normal(
        mean=center,
        cov=cov,
        size=n_samples_per_cluster
    )
    X.append(samples)
    y.append(np.full(n_samples_per_cluster, i))

X = np.vstack(X)
y = np.concatenate(y)

#  
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.scatter(
    np.array(centers)[:, 0],
    np.array(centers)[:, 1],
    c='black',
    marker='x',
    s=150,
    label='Centers'
)

plt.title('Overlapping Gaussian Clusters')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.show()
```
